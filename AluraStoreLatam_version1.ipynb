### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
# prompt: calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos

# Calculate the total revenue for each store
total_revenue_tienda1 = tienda['Precio'].sum()
total_revenue_tienda2 = tienda2['Precio'].sum()
total_revenue_tienda3 = tienda3['Precio'].sum()
total_revenue_tienda4 = tienda4['Precio'].sum()

print(f"Total revenue for Tienda 1: {total_revenue_tienda1}")
print(f"Total revenue for Tienda 2: {total_revenue_tienda2}")
print(f"Total revenue for Tienda 3: {total_revenue_tienda3}")
print(f"Total revenue for Tienda 4: {total_revenue_tienda4}")
```

# 2. Ventas por categoría

```python
# prompt: calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

# Function to analyze sales by category for a given store
def analyze_category_sales(df, store_name):
  # Group by category and count the number of sales
  category_sales = df['Producto'].value_counts().reset_index()
  category_sales.columns = ['Producto', 'Number of Sales']

  print(f"\nSales by category for {store_name}:")
  print(category_sales)

  # Find the most popular category
  most_popular_category = category_sales.iloc[category_sales['Number of Sales'].idxmax()]
  print(f"\nThe most popular category in {store_name} is: {most_popular_category['Producto']} with {most_popular_category['Number of Sales']} sales.")

# Analyze sales by category for each store
analyze_category_sales(tienda, 'Tienda 1')
analyze_category_sales(tienda2, 'Tienda 2')
analyze_category_sales(tienda3, 'Tienda 3')
analyze_category_sales(tienda4, 'Tienda 4')
```

```python
# prompt: con los resultados anteriores realiza un grafico donde quede claro que productos destavcaron en ventas en cada tienda

import matplotlib.pyplot as plt
import seaborn as sns

# Function to get the top N selling products for a store
def get_top_selling_products(df, n=5):
    product_sales = df['Producto'].value_counts().reset_index()
    product_sales.columns = ['Producto', 'Number of Sales']
    top_products = product_sales.head(n)
    return top_products

# Get the top 5 selling products for each store
top_products_tienda1 = get_top_selling_products(tienda)
top_products_tienda2 = get_top_selling_products(tienda2)
top_products_tienda3 = get_top_selling_products(tienda3)
top_products_tienda4 = get_top_selling_products(tienda4)

# Create subplots for each store
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Top 5 Selling Products by Store', fontsize=16)

# Plot for Tienda 1
sns.barplot(x='Number of Sales', y='Producto', data=top_products_tienda1, ax=axes[0, 0], palette='viridis')
axes[0, 0].set_title('Tienda 1')
axes[0, 0].set_xlabel('Number of Sales')
axes[0, 0].set_ylabel('Product')

# Plot for Tienda 2
sns.barplot(x='Number of Sales', y='Producto', data=top_products_tienda2, ax=axes[0, 1], palette='viridis')
axes[0, 1].set_title('Tienda 2')
axes[0, 1].set_xlabel('Number of Sales')
axes[0, 1].set_ylabel('Product')

# Plot for Tienda 3
sns.barplot(x='Number of Sales', y='Producto', data=top_products_tienda3, ax=axes[1, 0], palette='viridis')
axes[1, 0].set_title('Tienda 3')
axes[1, 0].set_xlabel('Number of Sales')
axes[1, 0].set_ylabel('Product')

# Plot for Tienda 4
sns.barplot(x='Number of Sales', y='Producto', data=top_products_tienda4, ax=axes[1, 1], palette='viridis')
axes[1, 1].set_title('Tienda 4')
axes[1, 1].set_xlabel('Number of Sales')
axes[1, 1].set_ylabel('Product')

plt.tight_layout(rect=[0, 0.03, 1, 0.95]) # Adjust layout to prevent title overlap
plt.show()
```

# 3. Calificación promedio de la tienda


```python
# prompt: calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.

# 3. Satisfacción del cliente por tienda

# Calculate the average customer rating for each store
average_rating_tienda1 = tienda['Calificación'].mean()
average_rating_tienda2 = tienda2['Calificación'].mean()
average_rating_tienda3 = tienda3['Calificación'].mean()
average_rating_tienda4 = tienda4['Calificación'].mean()

print(f"\nAverage customer rating for Tienda 1: {average_rating_tienda1:.2f}")
print(f"Average customer rating for Tienda 2: {average_rating_tienda2:.2f}")
print(f"Average customer rating for Tienda 3: {average_rating_tienda3:.2f}")
print(f"Average customer rating for Tienda 4: {average_rating_tienda4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# prompt: debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

import matplotlib.pyplot as plt
import seaborn as sns

def analyze_product_sales(df, store_name):
  # Group by product name and count the number of sales
  product_sales = df['Producto'].value_counts().reset_index()
  product_sales.columns = ['Producto', 'Number of Sales']

  print(f"\nProduct sales for {store_name}:")
  print(product_sales.head())
  print(product_sales.tail())

  # Find the most and least sold products
  most_sold_product = product_sales.iloc[product_sales['Number of Sales'].idxmax()]
  least_sold_product = product_sales.iloc[product_sales['Number of Sales'].idxmin()]

  print(f"\nThe most sold product in {store_name} is: {most_sold_product['Producto']} with {most_sold_product['Number of Sales']} sales.")
  print(f"The least sold product in {store_name} is: {least_sold_product['Producto']} with {least_sold_product['Number of Sales']} sales.")

  # Visualize the top 10 most sold products
  plt.figure(figsize=(12, 6))
  sns.barplot(x='Number of Sales', y='Producto', data=product_sales.head(10), palette='viridis')
  plt.title(f'Top 10 Most Sold Products in {store_name}')
  plt.xlabel('Number of Sales')
  plt.ylabel('Product Name')
  plt.show()

# Analyze product sales for each store
analyze_product_sales(tienda, 'Tienda 1')
analyze_product_sales(tienda2, 'Tienda 2')
analyze_product_sales(tienda3, 'Tienda 3')
analyze_product_sales(tienda4, 'Tienda 4')
```

# 5. Envío promedio por tienda

```python
# prompt: debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

# Calculate the average shipping cost for each store
average_shipping_tienda1 = tienda['Costo de envío'].mean()
average_shipping_tienda2 = tienda2['Costo de envío'].mean()
average_shipping_tienda3 = tienda3['Costo de envío'].mean()
average_shipping_tienda4 = tienda4['Costo de envío'].mean()

print(f"\nAverage shipping cost for Tienda 1: {average_shipping_tienda1:.2f}")
print(f"Average shipping cost for Tienda 2: {average_shipping_tienda2:.2f}")
print(f"Average shipping cost for Tienda 3: {average_shipping_tienda3:.2f}")
print(f"Average shipping cost for Tienda 4: {average_shipping_tienda4:.2f}")

# Create a bar plot to visualize the average shipping cost by store
stores = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
average_shipping_costs = [average_shipping_tienda1, average_shipping_tienda2, average_shipping_tienda3, average_shipping_tienda4]

plt.figure(figsize=(8, 6))
sns.barplot(x=stores, y=average_shipping_costs, palette='viridis')
plt.title('Average Shipping Cost by Store')
plt.xlabel('Store')
plt.ylabel('Average Shipping Cost')
plt.ylim(0, max(average_shipping_costs) * 1.1) # Adjust y-axis limit for better visualization
plt.show()
```
